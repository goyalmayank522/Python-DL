```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
cat59=cv2.imread('Cat_Dog_Dataset/train/CAT/59.jpg')
```

```python
cat59=cv2.cvtColor(cat59,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cat59)
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
image_gen = ImageDataGenerator(rotation_range=30,width_shift_range=0.1,height_shift_range=0.1,rescale=1/255,
                               shear_range=0.2,zoom_range=0.2,horizontal_flip=True,fill_mode='nearest')
```

```python
plt.imshow(image_gen.random_transform(cat59))
```

```python
image_gen.flow_from_directory('Cat_Dog_Dataset/train')
```

```python
from keras.models import Sequential
from keras.layers import Activation,Dropout,Flatten,Conv2D,MaxPooling2D,Dense
```

```python
model=Sequential()

model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(150,150,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(150,150,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(150,150,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(128))
model.add(Activation('relu'))

model.add(Dropout(0.5))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
batch_size=16

train_image_gen=image_gen.flow_from_directory('Cat_Dog_Dataset/train',target_size=(150,150),batch_size=batch_size,class_mode='binary')

test_image_gen=image_gen.flow_from_directory('Cat_Dog_Dataset/test',target_size=(150,150),batch_size=batch_size,class_mode='binary')
```

```python
train_image_gen.class_indices
```

```python
results=model.fit(x=train_image_gen,epochs=2,steps_per_epoch=150,validation_data=test_image_gen,validation_steps=12)
```

```python
results.history['accuracy']
# Less accuracy because the total number of epochs is very low i.e 2. So, if we increase the epochs to 100-150 then the accuracy will be quite better but the training time will be very high...
```
